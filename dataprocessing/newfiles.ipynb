```python
# create_sensor_folders.py：在每個包含 .csv 的資料夾下建立 FD、5D、B1、C8、D9 五個子資料夾

root_path = r"C:\Users\User\Desktop\高三專題\數據"
sensors = ['FD', '5D', 'B1', 'C8', 'D9']

def create_sensor_folders():
    # 取得所有包含 .csv 的資料夾路徑 (不重複)
    csv_files = glob.glob(os.path.join(root_path, "**/*.csv"), recursive=True)
    # 提取這些檔案所在的目錄路徑，並用 set 確保不重複處理同一個資料夾
    target_dirs = set(os.path.dirname(f) for f in csv_files)

    for directory in target_dirs:
        for s in sensors:
            # 拼出子資料夾路徑，例如: .../straight/1/FD
            sensor_dir = os.path.join(directory, s)
            
            os.makedirs(sensor_dir)

create_sensor_folders()
```

```python
root_path = r"C:\Users\User\Desktop\高三專題\數據"

def clean_original_csvs():
    all_files = glob.glob(os.path.join(root_path, "**/*g_out.csv"), recursive=True)
    
    deleted_count = 0
    
    for f in all_files:
        # 關鍵邏輯：只刪除「純 .csv」的原始檔，保留我們辛苦處理好的新檔
       
        os.remove(f)
        print(f"已刪除: {os.path.basename(f)}")
        deleted_count += 1
                
    print(f"\n清理完畢！共刪除了 {deleted_count} 個原始檔案。")

# 執行前請確認已備份重要數據
clean_original_csvs()
```
